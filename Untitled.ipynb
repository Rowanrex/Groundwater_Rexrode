#### Identify storm events for this location

```python

threshold = 0.1
philly_df_sorted = philly_df.sort_values('Date').reset_index(drop=True)
philly_df_sorted['Date'] = pd.to_datetime(philly_df_sorted['Date']).dt.normalize()

philly_df_sorted['is_storm_day'] = philly_df_sorted['ppt (inches)'] > threshold
philly_df_sorted['event_id'] = (philly_df_sorted['is_storm_day'] != philly_df_sorted['is_storm_day'].shift()).cumsum()

storm_events = philly_df_sorted[philly_df_sorted['is_storm_day']].groupby('event_id').agg(
    start_date=('Date', 'min'),
    end_date=('Date', 'max'),
    total_precip=('ppt (inches)', 'sum'),
    duration_days=('Date', lambda x: (x.max() - x.min()).days + 1)
).reset_index(drop=True)

print(f"Storm events identified: {len(storm_events)}")
storm_events.head()
```

#### Determine duration

```python

gw_sorted = gw_clean.sort_values('Date').reset_index(drop=True)

results = []

for _, storm in storm_events.iterrows():
    before = gw_sorted[gw_sorted['Date'] <= storm['start_date']]
    after = gw_sorted[gw_sorted['Date'] >= storm['end_date']]
    
    if len(before) == 0 or len(after) == 0:
        continue
    
    level_before = before.iloc[-1]
    level_after = after.iloc[0]
    
    gap_before = (storm['start_date'] - level_before['Date']).days
    gap_after = (level_after['Date'] - storm['end_date']).days
    
    results.append({
        'storm_start': storm['start_date'],
        'total_precip': storm['total_precip'],
        'duration_days': storm['duration_days'],
        'level_before': level_before['depth_to_water'],
        'level_after': level_after['depth_to_water'],
        'water_level_rise': level_before['depth_to_water'] - level_after['depth_to_water'],
        'gap_before_days': gap_before,
        'gap_after_days': gap_after
    })

response_df = pd.DataFrame(results)
print(response_df.shape)
print(response_df[['gap_before_days','gap_after_days']].describe())
```
